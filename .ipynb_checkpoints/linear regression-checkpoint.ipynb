```python
week_grouped_df = complete_df.groupby(['week'], as_index=False).mean().round(2)
```

```python
week_grouped_df.head(4)
```

```python
# get dummy variable for the holiday column
week_grouped_df = pd.get_dummies(week_grouped_df, columns = ['holiday'])
```

```python
week_grouped_df.columns.values
```

```python
# drop holiday_true column
week_grouped_df.drop(columns=['holiday_True'], inplace=True)
```

```python
week_grouped_df.head(2)
```

```python
# reorder columsn so the dependent variable is on the very right
# get rid of the store column
week_grouped_df = week_grouped_df.loc[:, week_grouped_df.columns[[0] + list(range(3,week_grouped_df.shape[1])) + [2]]]
```

```python
week_grouped_df.head(3)
```

```python
week_grouped_df.tail()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# does not include dates and holiday_False
predictors = week_grouped_df.columns.values[list(range(1,week_grouped_df.shape[1]-2))]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
week_grouped_df[predictors]
```

```python
x = week_grouped_df[predictors]
y = week_grouped_df['sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
yhat = lm.predict(x_test)
```

```python
yhat
```

```python
len(yhat)
```

```python
plt.figure(figsize=(15,10))
ax1 = sns.distplot(week_grouped_df['sales'], hist=False, color='r',label="Actual Data")

sns.distplot(yhat, hist=False, color="b", label = "Predicted Data", ax = ax1)
plt.savefig("distribution plot.jpg")
```

Measures of evaluation:

Determine how accurate the model is using R-squared and Mean Squared Error (MSE)



```python
# Find the R-squared using the fitted model from earlier
lm.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, yhat)
r2 = r2_score(y_test, yhat)
```
